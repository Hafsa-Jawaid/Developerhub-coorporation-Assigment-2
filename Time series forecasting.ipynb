```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
```

```python
# Simulate power consumption data
np.random.seed(42)
date_rng = pd.date_range(start='1/1/2022', end='1/1/2023', freq='H')
power_data = np.abs(np.random.normal(loc=1.5, scale=0.5, size=len(date_rng)))

df = pd.DataFrame({'datetime': date_rng, 'global_active_power': power_data})
df.set_index('datetime', inplace=True)
df_daily = df.resample('D').mean().reset_index()
```

```python
# Feature engineering
df_daily['day'] = df_daily['datetime'].dt.day
df_daily['month'] = df_daily['datetime'].dt.month
df_daily['weekday'] = df_daily['datetime'].dt.weekday
df_daily['dayofyear'] = df_daily['datetime'].dt.dayofyear
df_daily['week'] = df_daily['datetime'].dt.isocalendar().week.astype(int)
```

```python
# Prepare data for XGBoost
X = df_daily[['day', 'month', 'weekday', 'dayofyear', 'week']]
y = df_daily['global_active_power']
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)
```

```python
# ---- XGBoost Model ----
xgb = XGBRegressor()
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

```python
# ---- ARIMA Model ----
arima_model = ARIMA(df_daily['global_active_power'], order=(5,1,0))
arima_fit = arima_model.fit()
y_pred_arima = arima_fit.forecast(steps=len(X_test))
```

```python
# ---- Prophet Model ----
df_prophet = df_daily[['datetime', 'global_active_power']].rename(columns={'datetime': 'ds', 'global_active_power': 'y'})
prophet_model = Prophet()
prophet_model.fit(df_prophet)
future = prophet_model.make_future_dataframe(periods=len(X_test))
forecast = prophet_model.predict(future)
y_pred_prophet = forecast['yhat'][-len(X_test):].values
```

```python
# ---- Evaluation ----
def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"{model_name} - MAE: {mae:.4f}, RMSE: {rmse:.4f}")

evaluate_model(y_test, y_pred_xgb, "XGBoost")
evaluate_model(y_test, y_pred_arima, "ARIMA")
evaluate_model(y_test, y_pred_prophet, "Prophet")
```

```python
# ---- Plot Forecasts ----
plt.figure(figsize=(14, 5))
plt.plot(df_daily['datetime'][-len(X_test):], y_test.values, label='Actual', linewidth=2)
plt.plot(df_daily['datetime'][-len(X_test):], y_pred_xgb, label='XGBoost')
plt.plot(df_daily['datetime'][-len(X_test):], y_pred_arima, label='ARIMA')
plt.plot(df_daily['datetime'][-len(X_test):], y_pred_prophet, label='Prophet')
plt.title("Forecasted vs Actual Power Consumption")
plt.xlabel("Date")
plt.ylabel("Power (kW)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ---- Rolling Mean Plot ----
plt.figure(figsize=(12, 5))
plt.plot(df_daily['datetime'], df_daily['global_active_power'], label='Actual')
plt.plot(df_daily['datetime'], df_daily['global_active_power'].rolling(window=7).mean(), label='7-day Rolling Avg')
plt.title("Daily Power Consumption with Rolling Mean")
plt.xlabel("Date")
plt.ylabel("Power (kW)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ---- Residual Plot (XGBoost) ----
residuals = y_test - y_pred_xgb
plt.figure(figsize=(10, 4))
plt.hist(residuals, bins=30, color='teal', edgecolor='black')
plt.title("Residual Distribution (XGBoost Predictions)")
plt.xlabel("Error")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```
